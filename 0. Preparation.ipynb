# 1. Download and Run

## 1-1. Download the codelab materials

From [TensorFlow-Intro repository on GitHub](https://github.com/kazunori279/TensorFlow-Intro), click the green `Clone or download` button, download a zip file, and extract the zip file. You can move the extracted folder `TensorFlow-Intro-master` to anywhere you like. Then open [the top page of your local Datalab folder](http://localhost:8081/tree), navigate to the `TensorFlow-Intro-master` folder, so you will see the list of the codelab materials.
